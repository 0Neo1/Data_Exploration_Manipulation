### make fill missing values by - [SimpleImputer](https://scikit-learn.org/stable/modules/generated/sklearn.impute.SimpleImputer.html)

**NOTE: NOT DO NOT FILL OUTPUT COLUMN (y) ("RainToday" and "RainTomorrow")**

```python
from sklearn.impute import SimpleImputer                                        # load simple imputer
weather_df_filled_by_simpleImputer=weather_df.copy()                            # make copy of dataframe
#weather_df_filled_by_simpleImputer.dropna(subset=['RainToday','RainTomorrow'],inplace=True)# make drop rows having null in "RainToday" and "RainTomorrow"
#weather_df_filled_by_simpleImputer.isna().sum()                                 # make check (no null values in "RainToday" and "RainTomorrow")
# or better to drop "RainToday" and "RainTomorrow" columns, so that later on we can build machine learning model and predict missing values in this columns
# which is our actual task.
weather_df_filled_by_simpleImputer.drop(columns=['RainToday','RainTomorrow'],inplace=True)# make drop "RainToday" and "RainTomorrow"
pd.DataFrame(data={'null_count':weather_df_filled_by_simpleImputer.isna().sum(),# column wise null value count
                   'data_types':weather_df_filled_by_simpleImputer.dtypes})     # data type of each column
```

```python
# As Simple Imputer works on numerical or categorical column only so convert all object type columns to "category" type
# for column in weather_df_filled_by_simpleImputer.columns:                     # get all columns
#   if weather_df_filled_by_simpleImputer[column].dtype=='object':              # check column is object type
#     weather_df_filled_by_simpleImputer[column]=weather_df_filled_by_simpleImputer[column].astype('category')# make convert column to "category" type and overwite it dataframe copy
# direct method - get list of all object type (or other type) and overwite using loc or directly
object_type_columns=[column for column in weather_df_filled_by_simpleImputer.columns if weather_df_filled_by_simpleImputer[column].dtype=='object']# get all object type columns
#weather_df_filled_by_simpleImputer[object_type_columns]=weather_df_filled_by_simpleImputer[object_type_columns].astype('category')# make overwite "object" type columns to "category" type or
# using loc for columns access (object type only)
#weather_df_filled_by_simpleImputer.loc[:,object_type_columns]=weather_df_filled_by_simpleImputer.loc[:,object_type_columns].astype('category')# make overwite "object" type columns to "category" type
# `DeprecationWarning` by loc method
# or by pass dictionary to astype method
weather_df_filled_by_simpleImputer=weather_df_filled_by_simpleImputer.astype(dict(zip(object_type_columns,['category']*len(object_type_columns))))# make overwrite
weather_df_filled_by_simpleImputer.dtypes                                       # datatypes of each column
```

```python
# get list of all of categorical columns
categorical_columns=[column for column in weather_df_filled_by_simpleImputer.columns if weather_df_filled_by_simpleImputer[column].dtype=='category']
# get list of all of numerical columns
numerical_columns=[column for column in weather_df_filled_by_simpleImputer.columns if weather_df_filled_by_simpleImputer[column].dtype!='category']
```

```python
from sklearn.impute import SimpleImputer                                        # load simple imputer class
# make fill categorical column with the most occuring value - mode (column wise)
categoricalImputer=SimpleImputer(missing_values=np.nan,strategy='most_frequent')# fill categorical columns with 'most_frequent' (or most occuring) values
categoricalImputer.fit(weather_df_filled_by_simpleImputer[categorical_columns]) # make fit data (make internal tree)
# make overwrite all columns with fill value return by Simple Imputer object (categoricalImputer)
weather_df_filled_by_simpleImputer[categorical_columns]=categoricalImputer.transform(weather_df_filled_by_simpleImputer[categorical_columns])# make fill
# make fill numerical columns with mean of the data
numericalImputer=SimpleImputer(missing_values=np.nan,strategy='mean')           # fill numerical columns with 'mean' (or average) values of that column
# make overwrite all columns with fill value return by Simple Imputer object (numericalImputer)
weather_df_filled_by_simpleImputer[numerical_columns]=numericalImputer.fit_transform(weather_df_filled_by_simpleImputer[numerical_columns])# make fill
```

```python
weather_df_filled_by_simpleImputer.isna().sum().sum()                           # make print total null count (in filled data, check)
```

```python
# make plot all numerical columns and check before and after filling
number_of_points=1000                                                           # the number of points you want to downsample to
for column in numerical_columns:                                                # get a column from all numerical columns
  step_size=weather_df_filled_by_simpleImputer.shape[0]//number_of_points       # get step size
  # make plot downsampled orginal values (unfilled, orginal data) - 1000 values
  downsampled_unfilled=[pd.Series(weather_df[column][step:step+step_size]).mean() for step in range(0,weather_df.shape[0],step_size)]
  # make plot downsampled filled values (filled data by simple imputer)
  downsampled_filled=[pd.Series(weather_df_filled_by_simpleImputer[column][step:step+step_size]).mean() for step in range(0,weather_df_filled_by_simpleImputer.shape[0],step_size)]
  # make plot both data using line plot
  plt.figure(figsize=(15,7))                                                    # plot figure and set it's size
  plt.plot(downsampled_unfilled,color='red',linewidth=0.5,label='unfilled / orginal')# make plot orginal (unfilled) data
  plt.plot(downsampled_filled,color='green',alpha=0.3,label='filled')           # make plot filled data
  plt.grid(),plt.xlabel('index'),plt.ylabel(f'downsampled average value - {column}')# add grid and axis labels
  plt.title(f'{column}\'s filled vs unfilled (orginal) data'),plt.legend()      # add title and legend to plot
  plt.show()                                                                    # make show plot
```

```python
weather_df_filled_by_simpleImputer.to_csv('weather_df_filled_by_simpleImputer.csv',index=False)# make save filled dataframe
del weather_df_filled_by_simpleImputer                                          # make delete cleaned data
```

### make fill missing values by - [pandas.DataFrame.interpolate](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.interpolate.html)

Fill NaN values using an interpolation method (only numerical columns).

**NOTE: NOT DO NOT FILL OUTPUT COLUMN (y) ("RainToday" and "RainTomorrow")**

```python
weather_df_filled_by_interpolation=weather_df.copy()                            # make copy of dataframe
weather_df_filled_by_interpolation.drop(columns=['RainToday','RainTomorrow'],inplace=True)# make drop "RainToday" and "RainTomorrow"
pd.DataFrame(data={'null_count':weather_df_filled_by_interpolation.isna().sum(),# column wise null value count
                   'data_types':weather_df_filled_by_interpolation.dtypes})     # data type of each column
```

```python
# get list of all of categorical columns
categorical_columns=[column for column in weather_df_filled_by_interpolation.columns if weather_df_filled_by_interpolation[column].dtype=='object']
# get list of all of numerical columns
numerical_columns=[column for column in weather_df_filled_by_interpolation.columns if weather_df_filled_by_interpolation[column].dtype!='object']
```

```python
# make fill null values using interpolate - "linear" method - make overwrite
weather_df_filled_by_interpolation[numerical_columns]=weather_df_filled_by_interpolation[numerical_columns].interpolate(method='linear',axis=0)
# make check for null values in filled data
weather_df_filled_by_interpolation[numerical_columns].isna().sum()              # make check
# The reason why you are still seeing null values after applying cubic spline interpolation is likely due to the nature of the data and the
# interpolation method itself.
# Cubic spline interpolation works by fitting a piecewise cubic polynomial to the data points, which can introduce some level of error,
# especially in regions where the data is sparse or irregularly sampled. In such cases, the interpolation may not be able to accurately
# estimate the missing values.
```

```python
# make fill remining values using - `pandas.DataFrame.fillna` with mean value (or average value for numerical value left off by pandas.DataFrame.interpolate)
weather_df_filled_by_interpolation.fillna(value=weather_df_filled_by_interpolation[numerical_columns].mean(),inplace=True)# fill all numerical coulumns with their respective mean
# As pandas.DataFrame[column].mode() return pd.Series so take first most occuring value and make a column wise dictionary
categorical_columns_mode:dict={column:weather_df_filled_by_interpolation[column].mode()[0] for column in categorical_columns}# get column wise mode
print('Column wise mode:',categorical_columns_mode)                             # make print info
# fill all categorical coulumns using `pandas.DataFrame.fillna` with mode value (or most occuring value)
weather_df_filled_by_interpolation.fillna(value=categorical_columns_mode,inplace=True) # fill categorical columns with 'most_frequent' (or most occuring) values
weather_df_filled_by_interpolation.isna().sum().sum()                           # make print total null count (in filled data, check)
```

### make fill missing values by - [sklearn.impute.KNNImputer](https://scikit-learn.org/stable/modules/generated/sklearn.impute.KNNImputer.html)

**NOTE: NOT DO NOT FILL OUTPUT COLUMN (y) ("RainToday" and "RainTomorrow")**

**NOTE: Takes lots of time ~30 minutes**

```python
weather_df_filled_by_knn=weather_df.copy()                                      # make copy of dataframe
weather_df_filled_by_knn.drop(columns=['RainToday','RainTomorrow'],inplace=True)# make drop "RainToday" and "RainTomorrow"
pd.DataFrame(data={'null_count':weather_df_filled_by_knn.isna().sum(),          # column wise null value count
                   'data_types':weather_df_filled_by_knn.dtypes})               # data type of each column
```

```python
# get list of all categorical columns
categorical_columns=[column for column in weather_df_filled_by_knn.columns if weather_df_filled_by_knn[column].dtype=='object']
# get list of all numerical columns
numerical_columns=[column for column in weather_df_filled_by_knn.columns if weather_df_filled_by_knn[column].dtype!='object']
```

```python
from sklearn.preprocessing import LabelEncoder                                  # make load label encoder class object
# As KNN Imputer requires all values to be numerical so label encode all categorical columns
encoders=dict()                                                                 # make dictionary to store all encoder objects
for column in categorical_columns:                                              # get each column name from list of all categorical columns
  encoders[column]=LabelEncoder()                                               # make label encoder for each column (column_name (key) -> labelEncoder_object (value))
  weather_df_filled_by_knn[column]=encoders[column].fit_transform(weather_df_filled_by_knn[column])# make label encode all columns
  print('Labels / classes found in column -> {column}:',encoders[column].classes_)# make print info (classes)
```

```python
# make replace null value (label encoded) change back to null value -
# Steps -> Find the index of null values (for enocders dictionary) then then replace that (those) labels with numpy.nan
for column in categorical_columns:                                              # get each column name from list of all categorical columns
  # get each index (of nan value) and replace all null value index with np.nan
  weather_df_filled_by_knn[column].replace(to_replace=np.argwhere(pd.isna(encoders[column].classes_)).flatten(),value=np.nan,inplace=True)# make replace
# make check null values prpperly replaced ? Make check ---
assert all(weather_df_filled_by_knn.isna().sum()==weather_df.drop(columns=['RainToday','RainTomorrow']).isna().sum()),'Invaild null filled !!!!'
assert any(weather_df_filled_by_knn.dtypes!='object'),'Object type column find !'# make check. All must be numerical columns
weather_df_encoded=weather_df_filled_by_knn.copy()                              # make a copy of encoded unfilled data
```

```python
from sklearn.impute import KNNImputer                                           # make load KNN Imputer class
imputer=KNNImputer(missing_values=np.nan,n_neighbors=5,weights='distance')      # make build imputer
%time weather_numpy_array_filled_by_knn=imputer.fit_transform(weather_df_filled_by_knn)# fill all null values by KNN Imputer
```

```python
weather_df_filled_by_knn=pd.DataFrame(weather_numpy_array_filled_by_knn,columns=weather_df_filled_by_knn.columns)# make build data frame
weather_df_filled_by_knn.head()                                                 # see head of the filled data frame
```

```python
# make plot all numerical columns and check before and after filling
number_of_points=1000                                                           # the number of points you want to downsample to
for column in numerical_columns:                                                # get a column from all numerical columns
  step_size=weather_df_filled_by_knn.shape[0]//number_of_points       # get step size
  # make plot downsampled orginal values (unfilled, orginal data) - 1000 values
  downsampled_unfilled=[pd.Series(weather_df[column][step:step+step_size]).mean() for step in range(0,weather_df.shape[0],step_size)]
  # make plot downsampled filled values (filled data by simple imputer)
  downsampled_filled=[pd.Series(weather_df_filled_by_knn[column][step:step+step_size]).mean() for step in range(0,weather_df_filled_by_knn.shape[0],step_size)]
  # make plot both data using line plot
  plt.figure(figsize=(15,7))                                                    # plot figure and set it's size
  plt.plot(downsampled_unfilled,color='red',linewidth=0.5,label='unfilled / orginal')# make plot orginal (unfilled) data
  plt.plot(downsampled_filled,color='green',alpha=0.3,label='filled')           # make plot filled data
  plt.grid(),plt.xlabel('index'),plt.ylabel(f'downsampled average value - {column}')# add grid and axis labels
  plt.title(f'{column}\'s filled vs unfilled (orginal) data'),plt.legend()      # add title and legend to plot
  plt.show()                                                                    # make show plot
```

```python
# make plot all label encoded "categorical" columns before and after filling
for column in categorical_columns:                                              # get a column from all categorical columns
  plt.figure(figsize=(15,7))                                                    # plot figure and set it's size
  sns.histplot(data=pd.concat((                                                 # concat filled and unfilled data encoded columns
                               weather_df_encoded[[column]].rename(columns={column:column+'_unfilled'}),# make load unfilled column (encoded)
                               weather_df_filled_by_knn[[column]].rename(columns={column:column+'_filled'})# make load filled column (encoded)
                               ),axis=1),                                       # concatenate along columns
               kde=True,stat='percent',element='step')                          # make plot
  plt.grid(),plt.xlabel('index')                                                # add grid and axis labels
  plt.title(f'{column}\'s filled vs unfilled (orginal) data')                   # add title to plot
  plt.show()                                                                    # make show data
# Aggregate statistic to compute in each bin.
#   count: show the number of observations in each bin
#   frequency: show the number of observations divided by the bin width
#   probability or proportion: normalize such that bar heights sum to 1
#   percent: normalize such that bar heights sum to 100
#   density: normalize such that the total area of the histogram equals 1
# element : {"bars", "step", "poly"} Visual representation of the histogram statistic.
```
